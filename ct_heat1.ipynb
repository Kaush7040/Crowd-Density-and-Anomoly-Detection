```python
#working heatmap

import cv2
import numpy as np
from ultralytics import YOLO

# Path to your YOLOv8 model
model_path = r"C:\Users\Kaushik\code_JN\runs\detect\train2\weights\best.pt"

# Load the YOLOv8 model
model = YOLO(model_path)

# Open the webcam
cap = cv2.VideoCapture(0)  # 0 is the default webcam

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Run inference on the frame
    results = model(frame)

    # Get the number of bounding boxes
    num_boxes = len(results[0].boxes)

    # Extract bounding boxes
    centers = []
    for box in results[0].boxes.xyxy:  # Get bounding box coordinates
        x1, y1, x2, y2 = box[:4]  # Extract bounding box coordinates
        center_x = int((x1 + x2) / 2)
        center_y = int((y1 + y2) / 2)
        centers.append((center_x, center_y))

    # Create an empty heatmap
    heatmap = np.zeros((frame.shape[0], frame.shape[1]), dtype=np.float32)

    # Increase heat in areas where bounding box centers are close
    for center in centers:
        cv2.circle(heatmap, center, 50, (255), thickness=-1)  # Draw circles at detected centers

    # Apply Gaussian blur for a smooth heatmap
    heatmap = cv2.GaussianBlur(heatmap, (151, 151), 0)

    # Normalize the heatmap
    heatmap = cv2.normalize(heatmap, None, 0, 255, cv2.NORM_MINMAX)

    # Convert to color map
    heatmap_color = cv2.applyColorMap(heatmap.astype(np.uint8), cv2.COLORMAP_JET)

    # Overlay the heatmap onto the original frame
    overlay = cv2.addWeighted(frame, 0.7, heatmap_color, 0.5, 0)

    # Display the count of bounding boxes
    cv2.putText(
        overlay,
        f"Boxes: {num_boxes}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
    )

    # Show the frame with the heatmap
    cv2.imshow("YOLOv8 Heatmap", overlay)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# Path to your YOLOv8 model
model_path = r"C:\Users\Kaushik\code_JN\runs\detect\train2\weights\best.pt"

# Load the YOLOv8 model
model = YOLO(model_path)

# Open the webcam
cap = cv2.VideoCapture(0)  # 0 is the default webcam

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

# Create a persistent heatmap to accumulate movement
heatmap_accumulated = None
decay_factor = 0.997 # Controls how fast the trail fades (lower = fades slower)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Run inference on the frame
    results = model(frame)

    # Get the number of bounding boxes
    num_boxes = len(results[0].boxes)

    # Extract bounding box centers
    centers = []
    for box in results[0].boxes.xyxy:
        x1, y1, x2, y2 = box[:4]
        center_x = int((x1 + x2) / 2)
        center_y = int((y1 + y2) / 2)
        centers.append((center_x, center_y))

    # Initialize the heatmap if it's the first frame
    if heatmap_accumulated is None:
        heatmap_accumulated = np.zeros((frame.shape[0], frame.shape[1]), dtype=np.float32)

    # Apply decay to previous heatmap (fades out old positions)
    heatmap_accumulated *= decay_factor

    # Add new detections to the heatmap
    for center in centers:
        cv2.circle(heatmap_accumulated, center, 25, (255), thickness=-1)  # Smaller radius to focus heat

    # Apply Gaussian blur for smooth blending
    heatmap_blurred = cv2.GaussianBlur(heatmap_accumulated, (101, 101), 0)

    # Normalize for display
    heatmap_normalized = cv2.normalize(heatmap_blurred, None, 0, 255, cv2.NORM_MINMAX)

    # Convert to color map
    heatmap_color = cv2.applyColorMap(heatmap_normalized.astype(np.uint8), cv2.COLORMAP_JET)

    # Overlay heatmap onto the original frame
    overlay = cv2.addWeighted(frame, 0.7, heatmap_color, 0.5, 0)

    # Display the count of bounding boxes
    cv2.putText(
        overlay,
        f"Boxes: {num_boxes}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
    )

    # Show the frame with the heatmap
    cv2.imshow("YOLOv8 Heatmap with Trail", overlay)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
