```python
#running model + count + direction

import cv2
import torch
import numpy as np
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load YOLOv8 model
model_path = r"C:\Users\Kaushik\code_JN\runs\detect\train2\weights\best.pt"
model = YOLO(model_path)

# Initialize DeepSORT Tracker
tracker = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0)

# Open the webcam
cap = cv2.VideoCapture(0)  # 0 is the default webcam

if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

# Store previous positions of tracked objects for direction estimation
previous_positions = {}

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Run inference on the frame
    results = model(frame)

    # Get detected bounding boxes
    detections = results[0].boxes.xyxy.cpu().numpy()  # (x1, y1, x2, y2) format
    confidences = results[0].boxes.conf.cpu().numpy()  # Confidence scores

    # Filter detections (only process confident detections)
    min_confidence = 0.4  # Adjust as needed
    detections = [
        (tuple(box), conf, None)  # Format expected by DeepSORT
        for box, conf in zip(detections, confidences)
        if conf > min_confidence
    ]

    # Update DeepSORT tracker
    tracked_objects = tracker.update_tracks(detections, frame=frame)

    # Count the number of tracked objects
    num_boxes = len(tracked_objects)

    # Store direction text
    direction_text = "Unknown"

    for track in tracked_objects:
        if not track.is_confirmed():  
            continue  # Skip unconfirmed tracks

        track_id = track.track_id
        x1, y1, x2, y2 = track.to_tlbr()  # Get bounding box coordinates

        # Compute centroid
        centroid_x = int((x1 + x2) / 2)
        centroid_y = int((y1 + y2) / 2)

        # Determine movement direction
        if track_id in previous_positions:
            prev_x, prev_y = previous_positions[track_id]
            dx = centroid_x - prev_x
            dy = centroid_y - prev_y

            if abs(dx) > abs(dy):  
                if dx > 0:
                    direction_text = "Right"
                else:
                    direction_text = "Left"
            else:
                if dy > 0:
                    direction_text = "Down"
                else:
                    direction_text = "Up"
        
        # Update previous positions
        previous_positions[track_id] = (centroid_x, centroid_y)

        # Draw bounding box
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(frame, f"ID {track_id}", (int(x1), int(y1) - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Display the count and movement direction
    cv2.putText(frame, f"Count: {num_boxes}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f"Direction: {direction_text}", (10, 60),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)

    # Show the annotated frame
    cv2.imshow("YOLOv8 + DeepSORT", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

# Load YOLOv8 model
model_path = r"C:\Users\Kaushik\code_JN\runs\detect\train2\weights\best.pt"
model = YOLO(model_path)

# Initialize DeepSORT Tracker
tracker = DeepSort(max_age=30, n_init=3, nms_max_overlap=1.0)

# Open webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

# Store previous positions for direction estimation
previous_positions = {}
previous_angles = {}

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

    # Run YOLO inference on the frame
    results = model(frame)

    # Extract YOLO bounding boxes
    detections = results[0].boxes.xyxy.cpu().numpy()  # (x1, y1, x2, y2)
    confidences = results[0].boxes.conf.cpu().numpy()  # Confidence scores

    # Filter detections
    min_confidence = 0.4
    detections = [
        (tuple(box), conf, None) for box, conf in zip(detections, confidences) if conf > min_confidence
    ]

    # Update DeepSORT tracker
    tracked_objects = tracker.update_tracks(detections, frame=frame)

    # Count detected objects
    num_boxes = len(tracked_objects)

    # **1️⃣ Draw YOLO's original bounding boxes**
    for box in results[0].boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)  # Convert to integers
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box

    # **2️⃣ Draw movement arrows**
    for track in tracked_objects:
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        x1, y1, x2, y2 = track.to_tlbr()  # Bounding box from tracker

        # Compute centroid (center of bounding box)
        centroid_x = int((x1 + x2) / 2)
        centroid_y = int((y1 + y2) / 2)

        # Determine movement direction
        if track_id in previous_positions:
            prev_x, prev_y = previous_positions[track_id]
            dx = centroid_x - prev_x
            dy = centroid_y - prev_y

            if abs(dx) > 2 or abs(dy) > 2:  # Ignore tiny movements
                # Compute angle
                angle = np.arctan2(dy, dx)

                # Check if direction is stable
                if track_id in previous_angles:
                    prev_angle = previous_angles[track_id]
                    angle_change = abs(angle - prev_angle)

                    # Ignore erratic changes (> 45 degrees)
                    if angle_change < np.pi / 4:  
                        arrow_length = 30
                        arrow_end_x = int(centroid_x + arrow_length * np.cos(angle))
                        arrow_end_y = int(centroid_y + arrow_length * np.sin(angle))

                        # Draw a red arrow (indicating movement)
                        cv2.arrowedLine(frame, (centroid_x, centroid_y), (arrow_end_x, arrow_end_y),
                                        (0, 0, 255), 3, tipLength=0.3)

                        # Save stable angle
                        previous_angles[track_id] = angle
                else:
                    previous_angles[track_id] = angle  # Store first angle

        # Store previous position
        previous_positions[track_id] = (centroid_x, centroid_y)

    # **3️⃣ Display object count**
    cv2.putText(frame, f"Count: {num_boxes}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show frame
    cv2.imshow("YOLOv8 + DeepSORT Direction Estimation", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
